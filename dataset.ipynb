```python
gp=final_df.groupby('Vehicle')['Male'].sum().sort_values()
gp=gp.reset_index()
gp['Male']=gp['Male']/10000000
plt.figure(figsize=(12,8))
sn.lineplot(data=gp,x='Vehicle',y='Male')
plt.title('Men Population in Cr')
```

```python
gp=final_df.groupby('Vehicle')['Female'].sum().sort_values()
gp=gp.reset_index()
gp['Male']=gp['Female']/10000000
plt.figure(figsize=(12,8))
sn.lineplot(data=gp,x='Vehicle',y='Male')
plt.title('Men Population in Cr')
```

```python
dist=final_df[final_df['State name']=='uttar pradesh'].groupby('District name')['Population'].sum().reset_index()
dist['Population']=dist['Population']/10000000
sn.barplot(data=dist,x='District name',y='Population')
```

```python
dist = final_df[final_df['State name'] == 'uttar pradesh'].groupby('District name')['Population'].sum().reset_index()
dist['Population'] = dist['Population'] / 10000000
dist = dist.sort_values(by='Population')
ax = sn.barplot(data=dist, x='District name', y='Population')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.show()
```
